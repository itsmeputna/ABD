```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs4
```

```python

continents_page = requests.get("https://id.wikipedia.org/wiki/Pemilihan_umum_Presiden_Indonesia_2024").text
continents_page
```

```python
continents_countries_soup = bs4(continents_page,"lxml")
continents = continents_countries_soup.find_all('h2' > 'span', {"class":"mw-headline"})
continents
```
